```python
#Importing the libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#Importing the dataset

df = pd.read_csv('http://bit.ly/w-data')
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
#Splitting the dataset into the Training set and Test set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#Training the Simple Linear Regression model on the Training set

model = LinearRegression()
model.fit(X, y)
```

```python
#Predicting the Test set results

y_pred = model.predict(X_test)
```

```python
#What will be predicted score if a student studies for 9.25 hrs/ day
float(model.predict([[9.25]]))
```

```python
#Visualising the Training set results
plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
plt.style.use("seaborn-v0_8-whitegrid")
# Scatter plot
plt.scatter(X_train, y_train, color='red', label='Training data')

# Regression line
plt.plot(X_train, model.predict(X_train), color='blue', label='Regression line')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Hours vs Scores (Training set) ')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
plt.style.use("seaborn-v0_8-whitegrid")
# Scatter plot of the testing data
plt.scatter(X_test, y_test, color='red', label='Actual')

# Line plot of the predicted values
plt.plot(X_test, model.predict(X_test), color='blue', linewidth=2, label='Predicted')

# Add labels and title
plt.title('Hours vs Scores (Testing Set)')
plt.xlabel('Hours')
plt.ylabel('Scores')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test, y_pred)
print("Accuracy (R-squared score):", accuracy)
```
